### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('tips.csv')
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop(columns='Unnamed: 0', inplace=True)
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(tips['total_bill'],kde=False, bins=20)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(data=tips,x="tip", y="total_bill", hue="time")
plt.xlabel('tip')
plt.ylabel('total_bill')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.relplot(x=tips['tip'], y=tips['total_bill'], col=tips['size'],col_wrap=2)
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.scatterplot(x=tips['day'],y=tips['total_bill'])
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x=tips['tip'], y=tips['day'], hue=tips['sex'])
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(data=tips, x=tips['day'], y=tips['total_bill'],hue=tips['time'])
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
groupbytime = tips.groupby(tips['time'])
groupydiner = groupbytime.get_group('Dinner')
groupbylunch= groupbytime.get_group('Lunch')
plt.subplot(1,2,1)
plt.hist(groupydiner['tip'],bins=10,histtype='bar')
plt.ylabel('tip')
plt.xlabel('Dinner')
plt.subplot(1,2,2)
plt.hist(groupbylunch['tip'],bins=10,histtype='bar')
plt.ylabel('tip')
plt.xlabel('Lunch')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
group = tips.groupby('sex')
fig, axes = plt.subplots(1,2,sharex=False,sharey=False)
bymale = group.get_group('Male')
sns.scatterplot(ax=axes[0],x=bymale['sex'],y=tips['total_bill'],hue=tips['smoker'])
axes[0].set_title('BY MALE')
byfemale = group.get_group('Female')
sns.scatterplot(ax=axes[1],x=byfemale['sex'],y=tips['total_bill'],hue=tips['smoker'])
axes[1].set_title('BY FEMALE')

```
