### Step 4. How many products cost more than $10.00?

```python
chipko['item_price']=chipko['item_price'].str.replace('$','').astype(float)
chipko
filtered = chipko['item_price']>10
chipko.loc[filtered,['item_name']].count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
filtered = chipko['quantity'] ==1
filtere1 = chipko.loc[filtered,['item_name','item_price']]
filtered2 = filtere1.drop_duplicates(subset='item_name')
priceofeachitems = filtered2.reset_index().drop('index',axis=1)
priceofeachitems
```
